```python
import pandas as pd
import numpy as np
import colorsys

def flatten(dataset, label, key="colors"):
    rows = []
    for item in dataset:
        file_name = item.get("file")
        for c in item.get(key, []):
            rows.append({
                "dataset": label,
                "file": file_name,
                "hex": c["hex"],
                "percent": float(c["percent"])  # 0–100
            })
    return pd.DataFrame(rows)

df_fin = flatten(finland, "Finland")
df_mor = flatten(morocco, "Morocco")
df = pd.concat([df_fin, df_mor], ignore_index=True)

df.head()
```

```python
def hex_to_hsv(hex_code):
    h = hex_code.lstrip("#")
    r = int(h[0:2], 16)/255
    g = int(h[2:4], 16)/255
    b = int(h[4:6], 16)/255
    return colorsys.rgb_to_hsv(r, g, b)

hsv = df["hex"].apply(hex_to_hsv)
df["hue"] = hsv.apply(lambda x: x[0]*360)
df["sat"] = hsv.apply(lambda x: x[1])
df["val"] = hsv.apply(lambda x: x[2])
```

```python
def weighted_avg(x, w):
    return np.average(x, weights=w)

summary = (
    df.groupby("dataset")
      .apply(lambda g: pd.Series({
          "images": g["file"].nunique(),
          "neutral_ratio_%": weighted_avg((g["sat"] < 0.2).astype(float), g["percent"]) * 100,
          "weighted_saturation": weighted_avg(g["sat"], g["percent"]),
          "weighted_brightness": weighted_avg(g["val"], g["percent"]),
      }))
      .reset_index()
)

summary
```

```python
import numpy as np
import pandas as pd

def weighted_avg(x, w):
    return np.average(x, weights=w)

def compute_summary(df):
    summary = (
        df.groupby("dataset")
        .apply(lambda g: pd.Series({
            "Images": g["file"].nunique(),
            "Neutral_ratio_%": weighted_avg((g["sat"] < 0.2).astype(float), g["percent"]) * 100,
            "Weighted_saturation": weighted_avg(g["sat"], g["percent"]),
            "Weighted_brightness": weighted_avg(g["val"], g["percent"]),
            "Warm_hue_%": weighted_avg(((g["hue"] >= 0) & (g["hue"] <= 60)).astype(float), g["percent"]) * 100,
            "Cool_hue_%": weighted_avg(((g["hue"] >= 180) & (g["hue"] <= 260)).astype(float), g["percent"]) * 100,
        }))
        .reset_index()
    )
    return summary

summary = compute_summary(df)
summary
```

```python
import pandas as pd
import numpy as np
import colorsys

def rgb_to_hsv(r, g, b):
    # colorsys expects 0–1
    h, s, v = colorsys.rgb_to_hsv(r/255, g/255, b/255)
    hue_angle = h * 360
    return hue_angle, s, v

def is_warm_hue(h):
    # warm = reds/oranges/yellows + magenta side
    return (h <= 60) or (h >= 300)

def build_rows(items, climate_name):
    rows = []
    for img in items:
        for c in img["colors"]:
            r, g, b = c["rgb"]
            hue_angle, sat, bri = rgb_to_hsv(r, g, b)

            rows.append({
                "climate": climate_name,
                "file": img.get("file"),
                "path": img.get("path"),
                "rank": c.get("rank"),
                "hex": c.get("hex"),
                "r": r, "g": g, "b": b,
                "hue_angle": hue_angle,
                "saturation": sat,
                "brightness": bri,
                # percent is your weight (0–100)
                "weight": float(c.get("percent", 0.0)),
                "pixel_count": int(c.get("pixel_count", 0)),
            })
    return rows

rows = []
rows += build_rows(brazil, "brazil")
rows += build_rows(egypt, "egypt")
rows += build_rows(finland, "finland")

df = pd.DataFrame(rows)
df.head()
```

```python
NEUTRAL_SAT_CUTOFF = 0.20

df["is_neutral"] = df["saturation"] < NEUTRAL_SAT_CUTOFF
df["is_warm"] = df["hue_angle"].apply(is_warm_hue)
df["is_cool"] = ~df["is_warm"]
```

```python
def wmean(x, w):
    w = np.asarray(w, dtype=float)
    x = np.asarray(x, dtype=float)
    return np.sum(x * w) / np.sum(w)

def climate_summary(g):
    return pd.Series({
        "neutral_ratio_pct": 100 * wmean(g["is_neutral"].astype(float), g["weight"]),
        "weighted_saturation": wmean(g["saturation"], g["weight"]),
        "weighted_brightness": wmean(g["brightness"], g["weight"]),
        "warm_hue_pct": 100 * wmean(g["is_warm"].astype(float), g["weight"]),
        "cool_hue_pct": 100 * wmean(g["is_cool"].astype(float), g["weight"]),
        "n_color_rows": len(g),
        "n_images": g["file"].nunique(),
    })

summary = df.groupby("climate").apply(climate_summary)
summary
```

```python
import numpy as np

rng = np.random.default_rng(0)

def weighted_mean_from_rows(values, weights):
    return np.sum(values * weights) / np.sum(weights)

def perm_test_pairwise(df, climate_a, climate_b, col, n_perm=5000):
    A = df[df["climate"] == climate_a][[col, "weight"]].to_numpy(dtype=float)
    B = df[df["climate"] == climate_b][[col, "weight"]].to_numpy(dtype=float)

    vals = np.concatenate([A[:,0], B[:,0]])
    wts  = np.concatenate([A[:,1], B[:,1]])
    labels = np.array([0]*len(A) + [1]*len(B))

    # observed difference in weighted means
    obs = weighted_mean_from_rows(vals[labels==0], wts[labels==0]) - weighted_mean_from_rows(vals[labels==1], wts[labels==1])

    # permutation distribution
    count = 0
    for _ in range(n_perm):
        perm = rng.permutation(labels)
        diff = weighted_mean_from_rows(vals[perm==0], wts[perm==0]) - weighted_mean_from_rows(vals[perm==1], wts[perm==1])
        if abs(diff) >= abs(obs):
            count += 1

    p = (count + 1) / (n_perm + 1)
    return obs, p

pairs = [("brazil","egypt"), ("brazil","finland"), ("egypt","finland")]

for col in ["saturation", "brightness"]:
    print("\n===", col, "===")
    for a,b in pairs:
        obs, p = perm_test_pairwise(df, a, b, col, n_perm=5000)
        print(f"{a} vs {b}: observed diff={obs:.4f}, p={p:.4g}")
```

```python
for a,b in pairs:
    obs, p = perm_test_pairwise(df, a, b, "is_warm", n_perm=5000)
    print(f"{a} vs {b} warm-share diff={obs:.4f}, p={p:.4g}")
```

```python
for a,b in pairs:
    obs, p = perm_test_pairwise(df, a, b, "is_neutral", n_perm=5000)
    print(f"{a} vs {b} neutral-share diff={obs:.4f}, p={p:.4g}")
```
