```python
import colorsys

def flatten_results(data, country):
    rows = []
    for item in data:
        file = item.get("file")
        path = item.get("path")
        for c in item.get("colors", []):
            r, g, b = c["rgb"]
            # convert RGB (0–255) to HSV (0–1), then scale hue to degrees
            h, s, v = colorsys.rgb_to_hsv(r/255, g/255, b/255)
            rows.append({
                "country": country,
                "file": file,
                "path": path,
                "rank": c.get("rank"),
                "hex": c.get("hex"),
                "r": r, "g": g, "b": b,
                "percent": c.get("percent"),
                "pixel_count": c.get("pixel_count"),
                "hue_deg": h * 360,
                "sat": s,
                "val": v
            })
    return pd.DataFrame(rows)

df = pd.concat([
    flatten_results(brazil, "Brazil"),
    flatten_results(egypt, "Egypt"),
    flatten_results(finland, "Finland")
], ignore_index=True)

df.head(), df.shape
```

```python
# Per-image weighted metrics
def per_image_metrics(g):
    w = g["percent"].values
    sat_w = np.average(g["sat"].values, weights=w)
    val_w = np.average(g["val"].values, weights=w)
    cool_share = np.average(((g["hue"] >= 180) & (g["hue"] <= 260)).astype(float), weights=w)
    neutral_share = np.average((g["sat"] < 0.2).astype(float), weights=w)
    high_sat_share = np.average((g["sat"] > 0.6).astype(float), weights=w)

    return pd.Series({
        "sat_w": sat_w,
        "val_w": val_w,
        "cool_share": cool_share,
        "neutral_share": neutral_share,
        "high_sat_share": high_sat_share
    })

img_df = (
    df.groupby(["dataset","file"])
      .apply(per_image_metrics)
      .reset_index()
)

img_df.head()
```

```python
from scipy.stats import kruskal

def kruskal_test(metric):
    groups = [
        img_df[img_df["dataset"]=="Brazil"][metric],
        img_df[img_df["dataset"]=="Egypt"][metric],
        img_df[img_df["dataset"]=="Finland"][metric]
    ]
    stat, p = kruskal(*groups)
    return stat, p

metrics = ["sat_w", "cool_share", "neutral_share", "high_sat_share"]

for m in metrics:
    stat, p = kruskal_test(m)
    print(f"{m}: H={stat:.3f}, p={p:.6f}")
```

```python
import numpy as np

def permutation_test(metric, n_perm=5000):
    # observed group means
    groups = img_df.groupby("dataset")[metric]
    observed = groups.mean().max() - groups.mean().min()

    # pooled data
    pooled = img_df[metric].values
    labels = img_df["dataset"].values

    count = 0

    for _ in range(n_perm):
        shuffled = np.random.permutation(labels)
        temp = pd.DataFrame({"metric": pooled, "group": shuffled})
        diff = temp.groupby("group")["metric"].mean().max() - \
               temp.groupby("group")["metric"].mean().min()

        if diff >= observed:
            count += 1

    p_value = count / n_perm
    return observed, p_value
```

```python
metrics = ["sat_w", "cool_share", "neutral_share", "high_sat_share"]

for m in metrics:
    obs, p = permutation_test(m)
    print(f"{m}: observed diff = {obs:.4f}, p = {p:.5f}")
```

```python
import json
import pandas as pd
import numpy as np
import colorsys

# --- Load JSONs ---
with open("outputs/results_brazil.json") as f:
    brazil = json.load(f)
with open("outputs/results_egypt.json") as f:
    egypt = json.load(f)
with open("outputs/results_finland.json") as f:
    finland = json.load(f)

def flatten_full(dataset, label):
    rows = []
    for item in dataset:
        for c in item["colors"]:
            r, g, b = c["rgb"]
            h, s, v = colorsys.rgb_to_hsv(r/255, g/255, b/255)
            rows.append({
                "dataset": label,
                "file": item["file"],
                "rank_in_image": c.get("rank"),
                "hex": c["hex"],
                "r": r, "g": g, "b": b,
                "percent": float(c["percent"]),
                "hue_deg": h * 360,
                "sat": s,
                "val": v
            })
    return pd.DataFrame(rows)

df_full = pd.concat([
    flatten_full(brazil, "Brazil"),
    flatten_full(egypt, "Egypt"),
    flatten_full(finland, "Finland")
], ignore_index=True)

# --- Aggregate to get TOP 20 weighted per dataset ---
hex_totals = (
    df_full.groupby(["dataset", "hex", "r", "g", "b"], as_index=False)["percent"]
           .sum()
           .rename(columns={"percent": "total_weight"})
)

top_n = 20
top20 = (
    hex_totals.sort_values(["dataset", "total_weight"], ascending=[True, False])
              .groupby("dataset")
              .head(top_n)
              .copy()
)

# Add HSV for the aggregated colors
hsv = top20.apply(lambda row: colorsys.rgb_to_hsv(row["r"]/255, row["g"]/255, row["b"]/255), axis=1)
top20["hue_deg"] = hsv.apply(lambda x: x[0] * 360)
top20["sat"] = hsv.apply(lambda x: x[1])
top20["val"] = hsv.apply(lambda x: x[2])

# Add rank within dataset
top20["rank"] = top20.groupby("dataset")["total_weight"].rank(ascending=False, method="first").astype(int)

# Final columns (rounded for readability)
top20_out = top20[["dataset", "rank", "hex", "r", "g", "b", "hue_deg", "sat", "val", "total_weight"]].copy()
top20_out["hue_deg"] = top20_out["hue_deg"].round(2)
top20_out["sat"] = top20_out["sat"].round(3)
top20_out["val"] = top20_out["val"].round(3)
top20_out["total_weight"] = top20_out["total_weight"].round(2)

top20_out.sort_values(["dataset", "rank"])
```

```python
import json, pandas as pd, numpy as np, colorsys, math
from pathlib import Path


paths = {
    "brazil": "/Users/isabellaprimo/Downloads/TopColor-master/outputs/results_brazil.json",
    "egypt": "/Users/isabellaprimo/Downloads/TopColor-master/outputs/results_egypt.json",
    "finland": "/Users/isabellaprimo/Downloads/TopColor-master/outputs/results_finland.json",
}


def hex_to_rgb(hex_str):
    h = hex_str.lstrip("#")
    return tuple(int(h[i:i+2], 16) for i in (0,2,4))

def rgb_to_hsv_deg(rgb):
    r,g,b = rgb
    h,s,v = colorsys.rgb_to_hsv(r/255, g/255, b/255)
    return (h*360.0, s*100.0, v*100.0)

def rgb_to_hsl_deg(rgb):
    r,g,b = rgb
    h,l,s = colorsys.rgb_to_hls(r/255, g/255, b/255)  # note: HLS order
    return (h*360.0, s*100.0, l*100.0)

def warm_cool_from_hue(h):
    h = h % 360
    return "warm" if (h >= 330 or h < 90) else "cool"

def hue_bucket(h):
    h = h % 360
    buckets = [
        ("red", 345, 15), ("orange", 15, 45), ("yellow", 45, 75),
        ("yellow-green", 75, 105), ("green", 105, 135), ("green-cyan", 135, 165),
        ("cyan", 165, 195), ("cyan-blue", 195, 225), ("blue", 225, 255),
        ("blue-purple", 255, 285), ("purple", 285, 315), ("magenta", 315, 345),
    ]
    for name, start, end in buckets:
        if start < end:
            if start <= h < end: return name
        else:
            if h >= start or h < end: return name
    return "red"

rows = []
for climate, fp in paths.items():
    with open(fp, "r") as f:
        data = json.load(f)

    for item in data:
        img = item.get("file")
        img_path = item.get("path")
        for c in item.get("colors", []):
            hexv = c["hex"]
            rgb = tuple(c.get("rgb") or hex_to_rgb(hexv))
            pct = float(c.get("percent", np.nan))

            h, s, v = rgb_to_hsv_deg(rgb)
            _, _, l = rgb_to_hsl_deg(rgb)

            neutral = s < 12  # tweak this threshold if needed
            temp = "neutral" if neutral else warm_cool_from_hue(h)

            rows.append({
                "climate": climate,
                "image_file": img,
                "image_path": img_path,
                "rank_in_image": c.get("rank"),
                "hex": hexv,
                "r": rgb[0], "g": rgb[1], "b": rgb[2],
                "hue_deg": round(h, 2),
                "hue_bucket": hue_bucket(h),
                "sat_pct": round(s, 2),
                "val_pct": round(v, 2),
                "lightness_pct": round(l, 2),
                "temp_bin": temp,
                "is_neutral_low_sat": neutral,
                "percent_in_image": round(pct, 4),
                "pixel_count": c.get("pixel_count"),
            })

df = pd.DataFrame(rows)

def weighted_avg(group, col):
    w = group["percent_in_image"].to_numpy() / 100.0
    x = group[col].to_numpy()

    if col == "hue_deg":  # circular mean, ignoring neutrals
        mask = ~group["is_neutral_low_sat"].to_numpy()
        if mask.sum() == 0: return np.nan
        w = w[mask]
        h = x[mask] * np.pi/180.0
        sinm = np.sum(np.sin(h)*w)
        cosm = np.sum(np.cos(h)*w)
        ang = math.atan2(sinm, cosm)
        if ang < 0: ang += 2*math.pi
        return round(ang*180.0/np.pi, 2)

    wsum = w.sum()
    return np.nan if wsum == 0 else round(float(np.sum(x*w)/wsum), 2)

img_summary = (df.groupby(["climate","image_file","image_path"], as_index=False)
               .apply(lambda g: pd.Series({
                   "n_colors": len(g),
                   "warm_pct": round(float(g.loc[g["temp_bin"]=="warm","percent_in_image"].sum()), 3),
                   "cool_pct": round(float(g.loc[g["temp_bin"]=="cool","percent_in_image"].sum()), 3),
                   "neutral_pct": round(float(g.loc[g["temp_bin"]=="neutral","percent_in_image"].sum()), 3),
                   "avg_hue_deg": weighted_avg(g, "hue_deg"),
                   "avg_sat_pct": weighted_avg(g, "sat_pct"),
                   "avg_val_pct": weighted_avg(g, "val_pct"),
                   "avg_lightness_pct": weighted_avg(g, "lightness_pct"),
                   "dominant_hex": g.sort_values("percent_in_image", ascending=False).iloc[0]["hex"],
                   "dominant_pct": round(float(g["percent_in_image"].max()), 3),
               }))
               .reset_index(drop=True))

climate_summary = (img_summary.groupby("climate", as_index=False)
                   .agg(
                       n_images=("image_file","nunique"),
                       mean_warm_pct=("warm_pct","mean"),
                       mean_cool_pct=("cool_pct","mean"),
                       mean_neutral_pct=("neutral_pct","mean"),
                       mean_sat_pct=("avg_sat_pct","mean"),
                       mean_val_pct=("avg_val_pct","mean"),
                       mean_lightness_pct=("avg_lightness_pct","mean"),
                   )).round(2)

df.to_csv("topcolor_color_rows.csv", index=False)
img_summary.to_csv("topcolor_image_summary.csv", index=False)
climate_summary.to_csv("topcolor_climate_summary.csv", index=False)

display(df.head())
display(img_summary.head())
display(climate_summary)
```
